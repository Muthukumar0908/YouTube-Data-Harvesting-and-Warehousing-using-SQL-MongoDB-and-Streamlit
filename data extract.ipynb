```python
def create_table(sql):
    ##create channel_details
    mycursor.execute('''create table project.channel_details(
                channel_id varchar(255) primary key,
                channel_name varchar(255),
                description varchar(8000),
                subscribers int,
                view_count int,
                video_count int,
                publish_date varchar(255),
                play_list varchar(255))''')
    mydb.commit()
    #create playlist_details
    mycursor.execute('''create table project.playlist_details(
                play_list_id varchar(255) primary key,
                channel_id varchar(255) ,                
                play_list_name varchar(255),
                FOREIGN KEY (channel_id) REFERENCES channel_details(channel_id) ON UPDATE CASCADE )''')
    mydb.commit()
    #create video_details
    mycursor.execute('''create table project.video_details(
                video_id varchar(255) primary key,
                video_name varchar(500),
                channel_name varchar(255),
                video_description varchar(8000),
                video_view_count int,
                video_likes int,
                duration int,
                video_comment_count int,
                video_favorite_count  int,
                video_publishedAt varchar(255),
                thumbnails varchar (255),
                channel_id varchar(255),
                FOREIGN KEY (channel_id) REFERENCES channel_details(channel_id)  )''')
    mydb.commit()
    #create comment_details
    mycursor.execute('''create table project.comment_details(
                video_id varchar(255),
                comment_id varchar(255) primary key,
                comment_text varchar(500),
                comment_author varchar(255),
                comment_publishedAt varchar(255),
                FOREIGN KEY (video_id) REFERENCES video_details(video_id))''')
    mydb.commit()
create_table(all_de)    
```
